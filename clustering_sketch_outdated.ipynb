```python
import pandas as pd
data2017 = pd.read_stata("data/HCMST2017.dta")
data2017
```

```python
df_data2017 = data2017.dropna(axis='columns')
df_data2017
```

```python
#get all categorical columns
cat_columns = df_data2017.select_dtypes(['category']).columns

#convert all categorical columns to numeric
df_data2017[cat_columns] = df_data2017[cat_columns].apply(lambda x: pd.factorize(x)[0])

df_data2017
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df_kmeans = scaler.fit_transform(df_data2017)
```

```python
from sklearn.cluster import KMeans
kmeans_model = KMeans(n_clusters=4)
clusters = kmeans_model.fit_predict(df_data2017)
df_data2017.insert(df_data2017.columns.get_loc("CaseID"), "Cluster", clusters)
df_data2017.head(4)
```

```python
clusters
```

```python
import matplotlib.pyplot as plt
ssd = []
for k in range(2, 9):
    kmeans_model = KMeans(n_clusters=k)
    kmeans_model.fit(df_data2017)
    ssd.append(kmeans_model.inertia_)
plt.figure(figsize=(6, 4), dpi=100)
plt.plot(range(2, 9), ssd, color="green", marker="o")
plt.xlabel("Number of clusters (K)")
plt.ylabel("SSD for K")
plt.show()
```
